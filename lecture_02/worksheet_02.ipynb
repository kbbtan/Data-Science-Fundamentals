d) Complete the code below to replaces `male` with 0 and `female` with 1

```python
# Change to string 0s and 1s.
df["Sex"] = df["Sex"].replace("male", "0")
df["Sex"] = df["Sex"].replace("female", "1")

# Convert type to integers.
df["Sex"] = df["Sex"].astype(int)

df.head()
```

e) Complete the code below to add four columns `First Name`, `Middle Name`, `Last Name`, and `Title` corresponding to the value in the `name` column.

For example: `Braund, Mr. Owen Harris` would be: 

|First Name | Middle Name | Last Name | Title |
|-----------|-------------|-----------|-------|
| Owen      |  Harris     |  Braund   | Mr    |

Anything not clearly one of the above 4 categories can be ignored.

```python
def split_names(row):
    """ This function is applied row wise to the dataframe in order to extract fields from
        the name column.
    """
    values = row["Name"].split()
    name_fields = [None, None, None, None]
    
    # Extract different name portions from Name column.
    if len(values) >= 4:
        name_fields[1] = values[3]
        
    if len(values) >= 3:
        name_fields[0] = values[2]
        
    if len(values) >= 2:
        name_fields[3] = values[1][:-1]
        
    if len(values) >= 1:
        name_fields[2] = values[0][:-1]
    
    return name_fields

df[['First Name', 'Middle Name', 'Last Name', 'Title']] = df.apply(split_names, axis=1, result_type="expand")
df.head()
```

f) Complete the code below to replace all missing ages with the average age

```python
df['Age'] = df["Age"].fillna(df["Age"].mean())
df.head()
```

g) Plot a bar chart of the average age of those that survived and did not survive. Briefly comment on what you observe.
